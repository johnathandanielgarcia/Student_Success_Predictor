# Merging 2 Standardized Datasets

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
studentPerformance = pd.read_csv("/Users/johnathangarcia/Downloads/studentPerformancePredictor/data_exploration_analysis/NEW_Student_performance_data_.csv")
students2 = pd.read_csv("/Users/johnathangarcia/Downloads/studentPerformancePredictor/data_exploration_analysis/NEW_Students2.csv")
print(studentPerformance.head())
print(students2.head())
```

```python
merged = pd.merge(studentPerformance, students2, how='outer')
```

```python
merged
```

```python
merged = merged.fillna(0)
```

```python
merged
```

```python
plt.figure(figsize = (16, 12))
sns.heatmap(merged.corr(), annot = True, cmap = 'coolwarm')
```

```python
merged.to_csv("merged.csv", index = False)
```
